```python
import os
import torch
from torch.utils.data import DataLoader
from skimage.util import random_noise
import torchvision.transforms as transforms
from datasets.celeba import CelebA
import matplotlib.pyplot as plt
from PIL import Image
import torchvision as tv
from torchvision.transforms.functional import to_pil_image
```

```python
dataset = CelebA(root=os.path.join('exp/datasets'), split='test',
                 transform=transforms.Compose([
                     transforms.CenterCrop(140),
                     transforms.Resize(64),
                     transforms.ToTensor(),
                 ]), download=False)

dataloader = DataLoader(
    dataset,
    batch_size=1,
    shuffle=True,
    num_workers=0,
)
```

```python
samples, _ = next(iter(dataloader))
```

```python
samples.shape
```

```python
y_0_piosson = torch.from_numpy(random_noise(samples, mode='poisson'))
y_0_gaussian = torch.from_numpy(random_noise(samples, mode='gaussian'))
y_0_snp = torch.from_numpy(random_noise(samples, mode='s&p'))
y_0_salt = torch.from_numpy(random_noise(samples, mode='salt'))
y_0_pepper = torch.from_numpy(random_noise(samples, mode='pepper'))
y_0_speckle = torch.from_numpy(random_noise(samples, mode='speckle'))
```
